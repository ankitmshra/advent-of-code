```python
# load the inputs
inputs = []
with open("inputs/input.txt", 'r') as f:
    for i in f.read().split():
        inputs.append(i)
```

# Solution to First Question

```python
pos, cnt, pc = 50, 0, 0
for i in inputs:
    direction = i[:1]
    steps = int(i[1:])
    
    if direction == "L":
        steps = -steps
    
    pos = pos + steps
    if pos % 100 == 0:
        pc += 1
print(pc)
```

# Solution to 2nd Question

```python
pos = 50
pc = 0
cnt = 0

for i in inputs:
    direction = i[0]
    steps = int(i[1:])

    if direction == "L":
        # moving left: steps_to_first_zero is how many left steps needed so pos becomes 0
        # if pos == 0, the first hit is at 100 steps (i.e. after a full cycle)
        first = pos if pos != 0 else 100
        k = steps  # number of left steps
        if k >= first:
            pc += ( (k - first) // 100 ) + 1
        pos = (pos - k) % 100

    else:  # "R"
        # moving right: steps to first zero
        first = (100 - pos) if pos != 0 else 100
        s = steps  # number of right steps
        if s >= first:
            pc += ( (s - first) // 100 ) + 1
        pos = (pos + s) % 100

    print(f"{cnt}: {i}: {pos}: {pc}")
    cnt += 1

print("final pc =", pc)
```
