```python
def se(s, k):
    n = len(s)
    size = n // k
    return [s[i:i+size] for i in range(0, n, size)]

x = ["12121213","213213213", "111", "1111", "1111111", "1212121212", "12341234", "12311231", "99"]
c = 0
for i in x:
    n = len(i)
    div = 2
    for k in range(2, n+1):
        ls = se(i, k)
        if len(set(ls)) != 1:
            continue
        else:
            print(f"{i}: {ls}")
            break
        
        

print(c)
```
