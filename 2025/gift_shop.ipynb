```python
# load the data
with open("inputs/gift_shop.txt", "r") as file:
    content = file.read().strip()
    pids = content.split(",")

p = []
for pid in pids:
    start_int, end_int = pid.split("-")
    p.append((start_int, end_int))
print(p)
```

# 1st Part

```python

inv_sum = 0
for start, end in p:
    start = int(start)
    end = int(end)
    for i in range(start, end+1):
        istr = str(i)
        n = len(istr)
        if n % 2 == 0:
            if istr[:(n//2)] == istr[(n//2):]:
                inv_sum += i
print(inv_sum)
            
        
```

# 2nd Part

```python
def se(s, k):
    n = len(s)
    size = n // k
    return [s[i:i+size] for i in range(0, n, size)]

inv_sum = 0
for start, end in p:
    start = int(start)
    end = int(end)
    for i in range(start, end+1):
        istr = str(i)
        n = len(istr)
        for j in range(2, n+1):
            ls = se(istr, j)
            if len(set(ls)) != 1:
                continue
            else:
                print(f"{i}: {ls}")
                inv_sum += i
                break
            

print(inv_sum)
```
