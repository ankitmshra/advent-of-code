```python
with open("inputs/input.txt", "r") as f:
    contents = f.read().strip().splitlines()
```

```python
franges = []
iids = []
for i in contents:
    if "-" in i:
        rangetup = i.split("-")
        franges.append((int(rangetup[0]), int(rangetup[1])))
    else:
        try:
            iids.append(int(i))
        except:
            pass
```

```python
fresh_count = 0
for i in iids:
    for start, end in franges:
        if start <= i <= end:
            fresh_count += 1
            break
```

```python
print(fresh_count)
```

```python
franges.sort()
```

```python
merged = []
for start, end in franges:
    if not merged:
        merged.append([start, end])
    else:
        last_start, last_end = merged[-1]
        if start <= last_end + 1:
            merged[-1][1] = max(last_end, end)
        else:
            merged.append([start, end])

total_fresh_ids = 0
for start, end in merged:
    total_fresh_ids += end - start + 1

print(total_fresh_ids)
```
